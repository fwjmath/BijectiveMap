# Example 1: Star map

```python
def get_star(n):
    m = BijectiveMap()
    cornerlist = [1]
    for i in range(2*n - 1):
        cornerlist.append(m.add_vertex(0))
    return m
```

```python
mstar = get_star(5)
```

```python
mstar.plot()
```

```python
mstar.to_graph().plot()
```

# Example 2: Flower map

```python
def get_flower(n):
    m = BijectiveMap()
    cornerlist = [1]
    for i in range(2*n - 1):
        cornerlist.append(m.add_vertex(0))
    newclist = []
    for i in range(2*n):
        c = cornerlist.pop()
        newclist.append(m.add_vertex(c))
    for i in range(n):
        c1 = newclist.pop()
        c2 = newclist.pop()
        m.join_corners(c1, c2)
    return m
```

```python
get_flower(5).to_graph().plot()
```

```python
mflower=get_flower(3)
mflower.delete_edge(0)
mflower.to_graph().plot()
```

# Example 3: Dumbbel

```python
def get_dumbbell(n):
    m = BijectiveMap()
    cp = m.opposite_corner(0)
    cnew = m.add_vertex(0)
    for i in range(n):
        cnew = m.add_vertex(cnew)
    m.join_corners(0, cnew)
    cnew = m.add_vertex(cp)
    for i in range(n):
        cnew = m.add_vertex(cnew)
    m.join_corners(cp, cnew)
    return m
```

```python
mdumbbell = get_dumbbell(3)
```

```python
mdumbbell.plot()
```

```python
mdumbbell.to_graph().plot()
```
